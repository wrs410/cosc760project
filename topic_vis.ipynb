```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv('lda_VADERbmore.csv')
```

```python
df.head()
```

```python
#topics 0-5, 7, 8, 10-14 are the most coherent for this run
#topic 0 - weather
#topic 1 - COVID-19
#topic 2 - local politics
#Topic 3 - police, crime
#topic 4 - mid-term elections
#topic 5 - sports
#topic 6 -??
#topic 7 - Keith Davis trial
#topic 8 - restaurant openings, real-estate
#topic 9 - ??
#topic 10 - national politics
#topic 11 - local initiatives, organizations
#topic 12 - schools
#topic 13 - community activism, social justice
#topic 14 - local events
plt.figure()
pd.crosstab(df["Sentiment"],df["main_topic"],normalize="columns").transpose().plot.bar(stacked=True)
plt.title('Sentiment of tweets by topic')
plt.show()
```

```python
topic_num = df["main_topic"].value_counts()
```

```python
plt.figure()
ax = topic_num.plot(kind='bar')
ax.set_title("Topic")
ax.set_ylabel("Number of tweets")
plt.show()
```

```python
author_num = df["author"].value_counts()
```

```python
plt.figure()
ax = author_num.plot(kind='bar')
ax.set_title("Author")
ax.set_ylabel("Number of tweets")
plt.show()
```

```python
plt.figure()
pd.crosstab(df["Sentiment"],df["author"],normalize="columns").transpose().plot.bar(stacked=True)
plt.title('Sentiment of tweets by user')
plt.show()
```
